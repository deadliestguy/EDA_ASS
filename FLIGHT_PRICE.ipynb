```python
# date of jorney 
df['Date_of_Journey'].dtype
```

```python
# split the string 
df['day'] = df['Date_of_Journey'].str.split('/').str[0]
df['month'] =df['Date_of_Journey'].str.split('/').str[1]
df['year'] = df['Date_of_Journey'].str.split('/').str[2]
# but day month year str e ache ptake int e convert korte hbe 
```

```python
df.head(2)
```

```python
# alternative way 
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])
df['to_date']= df['Date_of_Journey'].dt.day
df['to_month']= df['Date_of_Journey'].dt.month
df['to_year']= df['Date_of_Journey'].dt.year
```

```python
df['to_date'].dtype
```

```python
# noiw delete the column date iof  jorney 
df.drop('Date_of_Journey' , axis = 1, inplace=True)
```

```python
df.dtypes
```

```python
# day month and year ke str to int e convert korte hbe 
df['day'] = df['day'].astype('int')
df['month'] = df['month'].astype('int')
df['year'] = df['year'].astype('int')
```

```python
df.dtypes
```

```python
# we can delete the column route 
df.drop('Route' , axis = 1,inplace = True)
df.head()
```

```python
# now we handeled dept_ time 
df['Dep_Time'].dtype  #string  type 22:30

# split the sept time 
df['dep_hr'] = df['Dep_Time'].str.split(':').str[0] #22
df['dep_min'] = df['Dep_Time'].str.split(':').str[1] #30

df.dtypes

# so we need to convert the column in int 
df['dep_hr'] = df['dep_hr'].astype('int')
df['dep_min'] = df['dep_min'].astype('int')

df.dtypes

# now dept time has no signifiance so we can delete the column 
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
# now its turn for arrival time 
df.dtypes  #object type  01:10 22 Mar 
df['Arrival_Time'].unique()

# split korbo 
df['arr_hr'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
df['arr_min'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]

df.head(2)

# convert it to int 
df['arr_hr'] = df['arr_hr'].astype('int')
df['arr_min'] = df['arr_min'].astype('int')

df.dtypes

# now Arrival_Time has no significance 
df.drop('Arrival_Time' , axis=1, inplace=True)

df.head(2)
```

```python
# now we have to hadneledd duration 
df['Duration'].unique() # 2h 30m #2h # 30m

# now split the data 
df['dur_hr'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['dur_min'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]

df.head()

# fil the nan 
df['dur_min'] = df['dur_min'].fillna(0)
df['dur_hr'] = df['dur_hr'].fillna(0)
df.head()

# 50m duration thakle ota hr er moddhe chole jabe 
```

```python
# now convert it to int
df['dur_hr'] = df['dur_hr'].astype('int')
df['dur_min'] = df['dur_min'].astype('int')

df.dtypes

# now dlt the duration colm 
df.drop('Duration',axis=1 , inplace = True)
```

```python
# ekhon total stops 
df['Total_Stops'].unique()
```

```python
# we have to handeled the missing value 
# mean , mdeian , mode 
# ekhane cat data - > mode 
df['Total_Stops'].mode()
# mode->1 stop
```

```python
# map korbo total stops 
# non stop =0  and baki guloke number wise
df['Total_Stops'].map({'non-stop': 0, 
                                           '1 stop' : 1,
                                           '2 stops' : 2,
                                           '3 stops' : 3,
                                           '4 stops' : 4,
                                            np.NaN   : 1})
```

```python
# cat data ->encoding 
# airline , souce and destiantion 
pd.Categorical(df['Airline'])
# 12 cat
pd.Categorical(df['Source'])
# sorc has 5 cat 
pd.Categorical(df['Destination'])
# 6 cat 
```

```python
# souce and destination is a cat data -> nominal 
# one hot encoding 
from sklearn.preprocessing import OneHotEncoder

# object crreation 
encoder = OneHotEncoder()

# fit and transform 
encoded_matrix = encoder.fit_transform(df[['Source','Destination']]) #mqtrix dei
# convert it to array 
encoded_arr = encoded_matrix.toarray()
# df e convert 
df1 = pd.DataFrame(encoded_arr,columns=encoder.get_feature_names_out())

# now dlt sorse and destination 
df.drop(['Source','Destination'] , axis=1,inplace=False)


```

```python
# concat df and df1 
pd.concat([df,df1])
```
