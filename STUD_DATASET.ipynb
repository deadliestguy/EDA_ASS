```python
df['total_score']=(df['math_score']+df['reading_score']+df['writing_score'])
df.head()
```

```python
# male/female kar avg besi check korbo with sub[lot 
plt.subplots(1,2,figsize=(10,4))
# 1st plot e akbo 
plt.subplot(121)
sns.histplot(data=df,x='avg_score',kde=True)
# conclusion - max bacchar avg 60-70 r modhhe 
# 2nd box e akbo 
plt.subplot(122)
sns.histplot(data=df,x='avg_score',kde=True, hue='gender')
# conlc-> female score better than males 
```
